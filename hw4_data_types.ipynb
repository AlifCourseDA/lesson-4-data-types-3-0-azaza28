Azbuka

```python
n = int(input())
flag = True
for i in range(n):
    if input().lower()[0] not in ('а', 'б', 'в'):
        flag = False
print("YES" if flag else 'NO')
```

Анонс новости

```python
length = int(input())
n = int(input())
ans = []
for i in range(n):
    if len(s := input()) <= length:
        ans.append(s)
    else:
        ans.append(s[:length - 3] + '...')
for i in ans:
    print(i)
```

Очистка данных


```python
ans = []
while s := input():
    if '##' == s[:2]:
        s = s[2:]
    if '@@@' == s[-3:]:
        continue
    ans.append(s)
for i in ans:
    print(i)
```

# bez kommentariev

```python
ans = []
while s := input():
    if '#' in s:
        ans.append(s[:s.index('#')])
    else:
        ans.append(s)
for i in ans:
    if len(i) != 0:
        print(i)
```

chastotniy analiz

```python
from collections import Counter
txt = ''
s = input().lower().replace(' ', '')
while s != 'финиш':
    txt += s
    s = input().lower().replace(' ', '')

d = sorted(Counter(txt).most_common())
print(max(d, key=lambda x: x[1])[0])
```

massovoe vozvedenie v stepen'

```python
n = int(input())
nums = list()
for i in range(n):
    nums.append(int(input()))
p = int(input())
for i in list((map(lambda x: x**p, nums))):
    print(i)
```

NOD

```python
from math import gcd
n = input()
n = list(map(int, n.split()))
g = n[0]
for i in range(1, len(n)):
    g = gcd(g, n[i])
print(g)
```

RLE

```python
s = input()
count = 1

for i in range(1, len(s)):
    if s[i] == s[i - 1]:
        count += 1
    else:
        print(s[i - 1], count)
        count = 1

print(s[-1], count)
```

## Symbolic difference

```python
a = [x for x in input()]
b = [x for x in input()]
txt = ''
for i in set.intersection(set(a), set(b)):
    txt += i
print(txt)
```

## kasheedi

```python
o = int(input())
m = int(input())
ovs = set()
man = set()
for i in range(o):
    ovs.add(input())
for i in range(m):
    man.add(input())
peresech = len(man.intersection(ovs))
print(peresech if peresech != 0 else 'Таких нет')
```

## Zayka pt2

```python
from collections import Counter
inp = list()
while s := input():
    inp.extend(s.split())
for name, cnt in Counter(inp).most_common():
    print(name, cnt)
```

## Transliteraciya

```python
txt = '''
А — A
Б — B
В — V
Г — G
Д — D
Е — E
Ё — E
Ж — Zh
З — Z
И — I
Й — I
К — K
Л — L
М — M
Н — N
О — O
П — P
Р — R
С — S
Т — T
У — U
Ф — F
Х — Kh
Ц — Tc
Ч — Ch
Ш — Sh
Щ — Shch
Ы — Y
Э — E
Ю — Iu
Я — Ia
'''

letters = dict()
txt = txt.replace('—', '').split()
keys = [txt[i] for i in range(0, len(txt), 2)]
values = [txt[i] for i in range(1, len(txt), 2)]

letters = {k: v for (k, v) in zip(keys, values)}
letters['Ь'] = ''
letters['Ъ'] = ''

res = ''
for i in input():
    if i.upper() == i and i in letters:
        res += letters.get(i, '')
    elif i.lower() == i and i.upper() in letters:
        res += letters.get(i.upper(), '').lower()
    else:
        res += i
print(res)
```

## Odnofamilci

```python
from collections import Counter
n = int(input())
inp = []
for i in range(n):
    inp.append(input())

flag = False
for surname, cnt in sorted(Counter(inp).most_common()):
    if cnt != 1:
        print(surname, '-', cnt)
        flag = True
if not flag:
    print("Однофамильцев нет")
```
